```python
#Analiz 1 – R&D Harcaması vs Kâr

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


# Veri yükleme
data = pd.read_csv('50_Startups.csv')
data.head()


# 1. GÖREV: R&D Harcaması vs Kâr (Scatter Plot)
num1 = data['R&D Spend']
num2 = data['Profit']
plt.figure(figsize=(8,5))
plt.scatter(num1, num2, color='blue')
plt.title("R&D Harcaması vs Kâr")
plt.xlabel("R&D Harcaması")
plt.ylabel("Kâr")
plt.grid(True)
plt.show()
```

- Grafiğin yatay ekseninde (x-ekseni) Ar-Ge Harcaması, dikey ekseninde (y-ekseni) ise Kâr yer alıyor.
- Noktaların genel olarak sağa doğru yükselen bir trend izlediği görülüyor. Bu, Ar-Ge harcamaları arttıkça kârın da genellikle arttığını gösteriyor.

Trend İlişkisi:
- Pozitif Korelasyon var gibi görünüyor: Ar-Ge harcamaları ile kâr arasında doğru orantılı bir ilişki mevcut.
- Yani, şirketler Ar-Ge'ye daha fazla yatırım yaptıkça, genellikle daha yüksek kâr elde ediyorlar.

```python
#Analiz 2 – Yönetim Harcaması vs Kâr

num1 = data['Administration']
num2 = data['Profit']
plt.figure(figsize=(8,5))
plt.scatter(num1, num2, color='green')
plt.title("Yönetim Harcaması vs Kâr")
plt.xlabel("Yönetim Harcaması")
plt.ylabel("Kâr")
plt.grid(True)
plt.show()
```

- Yatay eksen (x): Yönetim Harcaması
- Dikey eksen (y): Kâr
- Noktaların çoğu dikey bir şerit içinde toplanmış durumda, özellikle yönetim harcamalarının ≈75.000 - 95.000 aralığında yoğunlaştığı görülüyor.

Trend İlişkisi:
- Belirgin/Belirgin bir trend yok: Noktalar yaklaşık olarak dikey bir dağılım gösteriyor. Bu, yönetim harcamalarındaki değişikliklerle kâr arasında güçlü bir ilişki olmadığını gösteriyor.
- Aynı yönetim harcaması seviyesinde (örneğin 80.000) çok farklı kâr seviyeleri (100.000'den 180.000'e kadar) gözlemlenebiliyor.

```python
#Analiz 3 – Eyaletlere Göre Ortalama Kâr

ort = data.groupby('State')['Profit'].mean()
plt.figure(figsize=(8,5))
ort.plot(kind='bar', color='orange')
plt.title("Eyaletlere Göre Ortalama Kârlar")
plt.xlabel("Eyalet")
plt.ylabel("Ortalama Kâr")
plt.grid(axis='y')
plt.show()
```

- Dikey eksen: Ortalama Kâr (0-120.000 arası)
- Yatay eksen: Eyaletler (Florida, New York, California)

Eyaletlere Göre Kâr Karşılaştırması:
1. Florida:
- Ortalama kâr ≈ 115.000 civarında
- Üç eyalet arasında üst sırada yer alıyor
- En yüksek kâr performansı

2. New York:
- Ortalama kâr ≈ 110.000 civarında
- Üçü arasında ortalama performans

3. California:
- Ortalama kâr ≈ 105.000 civarında tahmin ediliyor
- En düşük kâr performansı


```python
#Analiz 4 – Harcama Dağılımları

bas = data[['R&D Spend', 'Administration', 'Marketing Spend']]
plt.figure(figsize=(8,5))
sns.boxplot(data=bas)
plt.title("Harcama Dağılımları")
plt.show()
```

- Kutu grafiği (box plot) formatında
- Üç ana harcama kalemi: R&D, Administration, Marketing
- Her kutu: dağılım, medyan, aykırı değerler hakkında bilgi veriyor

Bu dağılımdan hareketle şirketin:
- Teknoloji odaklı veya inovasyon merkezli bir yapıya sahip olduğu
- Büyüme aşamasında olabileceği (Ar-Ge ve pazarlamanın ön planda)
- Profesyonel yönetim yapısı bulunduğu sonucuna varılabilir

1. Marketing Spend (Pazarlama Harcamaları):

- Açık ara en yüksek harcama kalemi
- Grafikteki en uzun çubuk, şirketin pazarlamaya büyük önem verdiğini gösteriyor

2. R&D Spend (Ar-Ge Harcamaları):

- İkinci sırada yer alıyor
- Pazarlama harcamalarından belirgin şekilde düşük, ancak yönetim harcamalarından yüksek
- Şirketin inovasyona ve araştırma-geliştirmeye de kaynak ayırdığını gösteriyor

3. Administration (Yönetim Harcamaları):

- En düşük harcama kalemi
- Diğer iki kaleme kıyasla oldukça düşük seviyede
- Şirketin idari masrafları minimize ettiği anlaşılıyor
